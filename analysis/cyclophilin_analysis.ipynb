# Lets finally run some Random Forests

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

clf = RandomForestRegressor(n_estimators=150)
```

```python
test_set = pd.read_csv('cyclophilin.dat', sep=' ', index_col=False)
test_set
```

```python
target = np.array(test_set['A_score'])
target
```

```python
descriptors = np.array(test_set[['desolv_apol', 'desolv_pol', 'net_charge', 'mwt', 'logp', 'hbd', 'hba', 'rb']])
descriptors
```

```python
X_train, X_test, y_train, y_test = train_test_split(descriptors, target, test_size=0.5)
```

```python
len(X_train)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.n_features_
```

```python
test_results = clf.predict(X_test)
```

```python
test_results
```

```python
clf.score(descriptors, target)
```
